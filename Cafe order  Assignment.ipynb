```python
menu = {
    "tea": 120,
    "coffee": 200,
    "sandwich": 450,
    "cookie": 150,
    "cake_slice": 320
}

inventory = {
    "tea": 40,
    "coffee": 35,
    "sandwich": 20,
    "cookie": 50,
    "cake_slice": 25
}

orders = [
    {"tea": 2, "cookie": 1},
    {"coffee": 1, "sandwich": 2},
    {"tea": 3, "cake_slice": 1},
    {"coffee": 2, "cookie": 2, "sandwich": 1},
    {"tea": 1},
    {"coffee": 1, "cake_slice": 2}
]

print("STEP 1: DATA LOADED")
```

```python
processed_orders = []

for order in orders:
    cleaned_order = {}

    for item in order:
        # Item menu me nahi → ignore
        if item not in menu:
            continue

        req_qty = order[item]
        avail_qty = inventory[item]

        # Agar required quantity zyada hai → jitni available hai utni de do
        if req_qty > avail_qty:
            cleaned_order[item] = avail_qty
        else:
            cleaned_order[item] = req_qty

    processed_orders.append(cleaned_order)

print("STEP 2: PROCESSED ORDERS")
print(processed_orders)
```

```python
order_totals = []

for order in processed_orders:
    total = 0
    for item in order:
        total += order[item] * menu[item]
    order_totals.append(total)

print("STEP 3: ORDER TOTALS")
print(order_totals)
```

```python
items_sold = {
    "tea": 0,
    "coffee": 0,
    "sandwich": 0,
    "cookie": 0,
    "cake_slice": 0
}

for order in processed_orders:
    for item in order:
        qty = order[item]

        # Inventory update
        inventory[item] -= qty

        # Total items sold
        items_sold[item] += qty

print("STEP 4: UPDATED INVENTORY")
print(inventory)

print("\nITEMS SOLD TODAY:")
print(items_sold)
```

```python
# Total revenue (without discount)
total_revenue = sum(order_totals)

# Top selling item
top_selling_item = max(items_sold, key=items_sold.get)

print("STEP 5: DAY SUMMARY")
print("Total Revenue:", total_revenue)
print("Top Selling Item:", top_selling_item)
```

```python
low_stock = []

for item in inventory:
    if inventory[item] <= 5:
        low_stock.append(item)

print("STEP 6: LOW STOCK ITEMS (≤5)")
print(low_stock)
```

```python
order_totals_with_discount = []

for total in order_totals:
    if total > 1000:
        total = total - (total * 0.10)   # 10% discount
    order_totals_with_discount.append(total)

total_revenue_with_discount = sum(order_totals_with_discount)

print("STEP 7: ORDER TOTALS WITH DISCOUNT")
print(order_totals_with_discount)

print("Total Revenue After Discount:", total_revenue_with_discount)
```

```python
order_points = []
total_points = 0

for total in order_totals_with_discount:
    points = int(total // 100)   # 1 point per 100 PKR
    order_points.append(points)
    total_points += points

print("STEP 8: LOYALTY POINTS PER ORDER")
print(order_points)

print("TOTAL POINTS GIVEN TODAY:", total_points)
```
