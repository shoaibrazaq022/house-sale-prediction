####Data Cleaning

```python
df['Month - Year'] = df['Month - Year'].fillna(df['Month - Year'].mode()[0])
```

```python
df['Total Number of Accident'] = df['Total Number of Accident'].fillna(df['Total Number of Accident'].mean())
```

```python
df['Accident Fatal'] = df['Accident Fatal'].fillna(df['Accident Fatal'].mean())
```

```python
df['Accident Non-Fatal'] = df['Accident Non-Fatal'].fillna(df['Accident Non-Fatal'].mean())
```

```python
df['Person Killed'] = df['Person Killed'].fillna(df['Person Killed'].mean())
```

```python
df['Person Injured'] = df['Person Injured'].fillna(df['Person Injured'].mean())
```

```python
df['Total Number of vehicles Involves'] = df['Total Number of vehicles Involves'].fillna(df['Total Number of vehicles Involves'].mean())
```
