```python
# Step 1: Distance
speed = float(input("Enter speed in m/s: "))
time = float(input("Enter time in seconds: "))

distance = speed * time

print("Distance traveled:", distance, "meters")
50
```

```python
# Step 2: Force
mass = float(input("Enter mass in kg: "))
acceleration = float(input("Enter acceleration in m/s²: "))

force = mass * acceleration

print("Force applied:", force, "Newtons (N)")
```

```python
# Step 3: Gravitational Potential Energy
mass = float(input("Enter mass in kg: "))
height = float(input("Enter height in meters: "))

g = 9.8  # acceleration due to gravity
potential_energy = mass * g * height

print("Gravitational Potential Energy:", potential_energy, "Joules (J)")
```

```python
# Step 4: Kinetic Energy
mass = float(input("Enter mass in kg: "))
velocity = float(input("Enter velocity in m/s: "))

kinetic_energy = 0.5 * mass * velocity**2

print("Kinetic Energy:", kinetic_energy, "Joules (J)")
```

```python
# Step 5: Voltage using Ohm's Law
current = float(input("Enter current in Amperes (A): "))
resistance = float(input("Enter resistance in Ohms (Ω): "))

voltage = current * resistance

print("Voltage across resistor:", voltage, "Volts (V)")
```

```python
import math

# Step 6: Maximum Height
velocity = float(input("Enter initial velocity in m/s: "))
angle_deg = float(input("Enter launch angle in degrees: "))

g = 9.8  # gravity
angle_rad = math.radians(angle_deg)

max_height = (velocity**2 * math.sin(angle_rad)**2) / (2 * g)

print("Maximum height reached:", round(max_height, 2), "meters")
```

```python
import math

# Step 7: Period of a Spring
mass = float(input("Enter mass in kg: "))
spring_constant = float(input("Enter spring constant k in N/m: "))

period = 2 * math.pi * math.sqrt(mass / spring_constant)

print("Period of the spring-mass system:", round(period, 2), "seconds")
```

```python
import math

# Step 8: Free Fall Time
height = float(input("Enter height in meters: "))

g = 9.8  # gravity1
time_to_hit = math.sqrt(2 * height / g)

print("Time to hit the ground:", round(time_to_hit, 2), "seconds")
```

```python
import math

# Step 9: Work Done
force = float(input("Enter force in Newtons (N): "))
displacement = float(input("Enter displacement in meters: "))
angle_deg = float(input("Enter angle between force and displacement in degrees: "))

angle_rad = math.radians(angle_deg)
work_done = force * displacement * math.cos(angle_rad)

print("Work done by the force:", round(work_done, 2), "Joules (J)")
```

```python
# Step 10: Power Consumption
work = float(input("Enter work done in Joules (J): "))
time = float(input("Enter time in seconds: "))

power = work / time

print("Power consumed:", round(power, 2), "Watts (W)")
```
